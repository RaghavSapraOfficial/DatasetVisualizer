```python
import pandas as pd
# Function to determine if a column is continuous or categorical
def classify_variable(column):
    unique_values = data[column].nunique()
    if pd.api.types.is_numeric_dtype(data[column]):
        if unique_values > 20:
            return 'Continuous'
        else:
            return 'Categorical'
    else:
        return 'Categorical'

# Apply the function to each column
variable_types = {col: classify_variable(col) for col in data.columns}

# Print the variable types
for col, var_type in variable_types.items():
    print(f'{col}: {var_type}')
```

```python
import matplotlib.pyplot as plt
for column in data.columns:
    var_type = classify_variable(column)
    
    plt.figure(figsize=(20,15))
    
    if var_type == 'Categorical':
        data[column].value_counts().plot(kind='bar', color='skyblue')
        plt.title(f'Bar Graph of {column}')
        plt.xlabel(column)
        plt.ylabel('Count')
    else:
        data[column].plot(kind='hist', bins=30, color='skyblue')
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
    
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```
